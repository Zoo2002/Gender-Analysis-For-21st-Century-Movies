## MOVIES INFO

```python
movies_data=pd.read_csv('/content/drive/MyDrive/PROJECT/movie_screenplays/movies_tmdb_id.csv')
```

```python
movies_data
```

```python
for idx, row in movies_data.iterrows():
  movie_id = row['MOVIE_ID']

  try:

    movie_details = movie_api.details(movie_id)
    credits = movie_api.credits(movie_id)

    movies_data.loc[idx,'GENRES'] = ', '.join([g.name for g in movie_details.genres]) if movie_details.genres else None
    movies_data.loc[idx,'OVERVIEW'] = movie_details.overview
    movies_data.loc[idx,'RUNTIME [MIN]'] = movie_details.runtime
    movies_data.loc[idx,'RATING'] = movie_details.vote_average
    movies_data.loc[idx,'BUDGET'] = movie_details.budget
    movies_data.loc[idx,'REVENUE'] = movie_details.revenue

    for person in credits.crew:
      if person.job == 'Director':
        director_name = person.name

        try:
          director_details = person_api.details(person.id)
          gender_map = {0: "Unknown", 1: "Woman", 2: "Man", 3: "Other"}
          director_gender = gender_map.get(director_details.gender, "Unknown")
        except Exception:
          director_gender = "Unknown"

        movies_data.loc[idx,'DIRECTOR'] = director_name
        movies_data.loc[idx,'DIRECTOR_GENDER'] = director_gender
        break

  except Exception as e:
    print(f"Błąd dla MOVIE_ID = {movie_id}: {e}")
    continue

```

## PEOPLE INFO

```python
movies_data=pd.read_csv('/content/drive/MyDrive/PROJECT/movie_screenplays/movies_tmdb_id.csv')
```

```python
movies_data
```

```python
actors_data_list = []

for idx, row in movies_data.iterrows():

    movie_id     = row["MOVIE_ID"]
    release_year = row["YEAR"]

    try:
        credits = movie_api.credits(movie_id)
    except Exception as e:
        print(f"ERROR credits failed - MOVIE_ID={movie_id} - {e}")
        continue

    cast_list = list(credits.cast)
    for actor in cast_list[:20]:

        try:
            person_details = person_api.details(actor.id)
        except Exception as e:
            print(f"ERROR person details failed - {actor} - {e}")
            continue

        gender_map = {0: "Unknown", 1: "Woman", 2: "Man", 3: "Other"}
        gender = gender_map.get(getattr(person_details, "gender", 0), "Unknown")

        if getattr(person_details, "birthday", None):
            birth_year = int(person_details.birthday[:4])
            age = release_year - birth_year
        else:
            age = "Unknown"

        try:
            actors_data_list.append({
                "TITLE": row["TITLE"],
                "YEAR": release_year,
                "MOVIE_ID": movie_id,
                "ACTOR_NAME": actor.name,
                "CHARACTER": actor.character,
                "GENDER": gender,
                "AGE_AT_RELEASE": age,
                "ACTOR_ID": actor.id,})

        except Exception as e:
            print(f" ERROR append failed - {actor.id} - {e}")
            continue
```

```python
actors_data = pd.DataFrame(actors_data_list)
```

```python
actors_data
```

```python
actors_data.to_csv('/content/drive/MyDrive/PROJECT/movie_screenplays/actors_tmdb_details.csv', index=False, encoding="utf-8-sig")
```

```python
from genderize import Genderize
```

```python
g = Genderize()
for id, row in actors_data.iterrows():
  if row['GENDER']=='Unknown':
    name = row['ACTOR_NAME'].split()[0]
    gender_res = g.get([name])[0]['gender']
    prob_res = g.get([name])[0]['probability']
    if prob_res > 0.9:
      if gender_res == 'male':
        actors_data.loc[id, 'GENDER'] = 'Man'
      if gender_res == 'female':
        actors_data.loc[id, 'GENDER'] = 'Woman'
```
