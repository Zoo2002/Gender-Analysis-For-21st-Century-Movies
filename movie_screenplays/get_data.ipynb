```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
from rapidfuzz import fuzz
```

```python
headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:142.0) Gecko/20100101 Firefox/142.0"}
```

```python
movies=[]
for year in range(2000,2026):
  counter=0
  print(f'year:{year} - DONE')
  url=f'https://www.boxofficemojo.com/year/world/{year}/'
  response=requests.get(url, headers=headers)
  response.raise_for_status()
  response.encoding='utf-8'
  html=response.text
  soup=BeautifulSoup(html, "html.parser")
  title_tags=soup.select("table a.a-link-normal")
  title_tags=title_tags[6:]
  for title_tag in title_tags:
    title=title_tag.get_text(strip=True)
    if title not in [movie['title'] for movie in movies]:
      movies.append({'year':year,'title':title})
      counter+=1
      if counter==120:
        break
```

```python
movies_df=pd.DataFrame(movies)
```

```python
movies_df
```

```python
movies_df=pd.DataFrame({'ID': range(0,len(movies_df)), 'YEAR': movies_df['year'], 'TITLE': movies_df['title'], 'SOURCE': 'box_office'})
```

Rolling Stone

```python
movies_rolling_stone=[]
url1='https://www.rollingstone.com/tv-movies/tv-movie-lists/best-movies-of-the-21st-century-1235200512/'
url2='https://www.rollingstone.com/tv-movies/tv-movie-lists/best-movies-of-the-21st-century-1235200512/hunger-2008-1235354192/'
urls=[url1,url2]
for url in urls:
  response=requests.get(url, headers=headers)
  response.raise_for_status()
  response.encoding='utf-8'
  html=response.text
  soup=BeautifulSoup(html, "html.parser")
  title_tags=soup.select("h2")[0:50]
  for title_tag in title_tags:
    title_year=title_tag.get_text(strip=True)
    title=re.search(r"[‘’](.+?)’", title_year).group(1)
    try:
      year=re.search(r"\((\d{4})\)", title_year).group(1)
    except:
      year=re.search(r"\((\d{4}-\d{4})\)", title_year).group(1)
    movies_rolling_stone.append({'year':year,'title':title})

```

Rotten Tomatoes

```python
movies_rott_tom=[]
url='https://editorial.rottentomatoes.com/guide/best-movies-21st-century/'
response=requests.get(url, headers=headers)
response.raise_for_status()
response.encoding='utf-8'
html=response.text
soup=BeautifulSoup(html, "html.parser")
h2_tags=soup.select("h2")
for h2_tag in h2_tags:
  title_tag=h2_tag.find('a')
  year_tag=h2_tag.find('span', class_='subtle start-year')
  if title_tag and year_tag:
    title=title_tag.get_text(strip=True)
    year=year_tag.get_text(strip=True).strip('()')
    movies_rott_tom.append({'title':title,'year':year})
```

```python
movies_guardian=[]
url='https://www.theguardian.com/film/2019/sep/13/100-best-films-movies-of-the-21st-century'
response=requests.get(url, headers=headers)
response.raise_for_status()
response.encoding='utf-8'
html=response.text
soup=BeautifulSoup(html, "html.parser")
title_tags=soup.select("h2")#.get_text(strip=True)
for title_tag in title_tags:
    strong_tag=title_tag.find('strong')
    if strong_tag:
        title=strong_tag.get_text(strip=True)
        year_match=re.search(r"\((\d{4})\)", str(title_tag))
        if year_match:
            year=year_match.group(1)
            movies_guardian.append({'title': title, 'year': year})
        else:
            print("Error in tag:", title_tag)
```

```python
base_link = 'https://www.scriptslug.com/script/'
save_dir = '/content/drive/MyDrive/PROJECT/movie_screenplays/raw_screenplays'

for idx, row in movies_filtered_df.iterrows():

    title = row["TITLE"]
    year = row["YEAR"]

    title_format=slugify(title)
    slug_with_year=f"{title_format}-{year}"
    url=f"{base_link}{slug_with_year}"

    try:
        head_response = requests.head(url, allow_redirects=True, timeout=5)

        if head_response.status_code != 200:
            movies_filtered_df.loc[idx, "STATUS"] = "unavailable"
            movies_filtered_df.loc[idx, "SCREENPLAY_SOURCE"] = None
            print(f"{idx}, {title}: UNAVAILABLE")
            continue

        page_response = requests.get(url, headers=headers, timeout=10)
        page_response.raise_for_status()

        soup = BeautifulSoup(page_response.text, "html.parser")
        pdf_tag = soup.find("a", class_="js-button-read")

        if not pdf_tag or "href" not in pdf_tag.attrs:
            movies_filtered_df.loc[idx, "STATUS"] = "no_pdf_link"
            movies_filtered_df.loc[idx, "SCREENPLAY_SOURCE"] = None
            print(f"{idx}, {title}: NO PDF LINK")
            continue

        pdf_url = pdf_tag["href"]
        pdf_response = requests.get(pdf_url, headers=headers, timeout=15)
        pdf_response.raise_for_status()

        filename = f"{title}.pdf"
        filepath = os.path.join(save_dir, filename)

        with open(filepath, "wb") as f:
            f.write(pdf_response.content)


        movies_filtered_df.loc[idx, "STATUS"] = "downloaded"
        movies_filtered_df.loc[idx, "SCREENPLAY_SOURCE"] = "slug"

        print(f"{idx}, {title}: DOWNLOADED")

    except requests.RequestException:
        movies_filtered_df.loc[idx, "STATUS"] = "unavailable"
        movies_filtered_df.loc[idx, "SCREENPLAY_SOURCE"] = None
        print(f"{idx}, {title}: ERROR")
```
